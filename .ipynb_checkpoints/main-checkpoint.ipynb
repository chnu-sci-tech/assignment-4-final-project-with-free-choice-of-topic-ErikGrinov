# T-Test for Difference in Means

UCI Heart Disease Data

Dataset for analyzing heart disease risk.
Patients are divided by gender, age (young and old), presence or absence of heart disease.
You can investigate which groups have a higher risk of the disease.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the data
data = pd.read_csv('resources/heart_disease_uci.csv')

# Handle missing values (if any)
data = data.dropna()

# Preprocess the 'age' column into 'young' and 'old' categories
age_threshold = 45
data['age_group'] = data['age'].apply(lambda x: 'young' if x <= age_threshold else 'old')

# Convert 'sex' and 'dataset' to categorical (if needed)
data['sex'] = data['sex'].map({'Male': 1, 'Female': 0})

# Perform T-test for cholesterol between male and female
male_chol = data[data['sex'] == 1]['chol']
female_chol = data[data['sex'] == 0]['chol']

t_stat, p_value = stats.ttest_ind(male_chol, female_chol)
print(f"T-test for cholesterol between male and female: t-stat = {t_stat}, p-value = {p_value}")

# Perform T-test for cholesterol between young and old
young_chol = data[data['age_group'] == 'young']['chol']
old_chol = data[data['age_group'] == 'old']['chol']

t_stat_age, p_value_age = stats.ttest_ind(young_chol, old_chol)
print(f"T-test for cholesterol between young and old: t-stat = {t_stat_age}, p-value = {p_value_age}")

# Perform T-test for cholesterol between patients with and without heart disease
disease_chol = data[data['num'] == 1]['chol']
no_disease_chol = data[data['num'] == 0]['chol']

t_stat_disease, p_value_disease = stats.ttest_ind(disease_chol, no_disease_chol)
print(f"T-test for cholesterol between patients with and without heart disease: t-stat = {t_stat_disease}, p-value = {p_value_disease}")

# Visualization with hue argument to avoid warning
plt.figure(figsize=(12, 6))
sns.boxplot(x='sex', y='chol', data=data, hue='sex')
plt.title('Cholesterol Distribution by Sex')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='age_group', y='chol', data=data, hue='age_group')  # Using hue to define grouping
plt.title('Cholesterol Distribution by Age Group')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='num', y='chol', data=data, hue='num')  # Using hue to define grouping
plt.title('Cholesterol Distribution by Heart Disease Status')
plt.show()

```

Detailed interpretation of the results shown in the charts and statistical tests for cholesterol levels:

1. Cholesterol Distribution by Sex
   
Description: The boxplot shows cholesterol levels by sex, where 0 represents females and 1 represents males.

Interpretation: The median cholesterol level appears slightly higher for females than males, though there’s substantial overlap in the interquartile range.

T-test Result: The t-statistic is approximately -1.36, and the p-value is around 0.1738. Since the p-value is greater than 0.05, this suggests that there is no statistically significant difference in cholesterol levels between males and females.

2. Cholesterol Distribution by Age Group
   
Description: This boxplot illustrates cholesterol levels by age group, with young defined as those aged 45 or below and old as those above 55.

Interpretation: Both age groups have similar median cholesterol levels, with no major visual differences in their distributions.

T-test Result: The t-statistic is about -0.55, and the p-value is approximately 0.5837. Again, the p-value is above 0.05, suggesting no significant difference in cholesterol levels between the young and old groups.

3. Cholesterol Distribution by Heart Disease Status
   
Description: This boxplot shows cholesterol levels according to heart disease status, where 0 represents no heart disease, and higher numbers (up to 4) indicate the presence of heart disease, with 4 being the most severe.

Interpretation: Cholesterol levels seem to have minor variations across heart disease severity levels, but these differences are not visually pronounced.

T-test Result: The t-statistic is approximately -1.78, and the p-value is around 0.0756. With a p-value just above 0.05, there is no statistically significant difference in cholesterol levels between individuals with and without heart disease, although it is close to being significant.

Summary

The t-tests indicate that there are no statistically significant differences in cholesterol levels across sex, age group, or heart disease status. This suggests that in this dataset, cholesterol levels are relatively similar across these demographic and health categories.
